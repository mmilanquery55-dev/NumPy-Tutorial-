Array Stacking  and Spliting

```python
# Stacking -> 1.Vertical stacking , 2. Horizontal Stacking
a=np.array([1,2,3])
b=np.array([4,5,6])

print(np.vstack((a,b)))
print(np.hstack((a,b)))

```

```python
#  Split -> in simple term split is using for breaking the array
c=np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
print(np.hsplit(c,3))
```

```python
# Horizontal Split
split=np.hsplit(c,3)
for s in split:
  print(s)
```

```python
# Vertical Split
d=np.array([[1,2,3],[4,5,6]])
print(d)
```

```python
split=np.vsplit(d,2)
print(split)
```

```python
split = np.vsplit(d,2)
for s in split:
  print(s)
```

Mathematical Operations

Mathematical Operations on Multiple Arrays

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
print(np.mod(a,b))
```

```python
# dot product -> fistly multipling array a and array b each
# other and then adding the total value
# which is comes in multiply a and b lastly show the value

print(np.dot(a,b))
```

```python
# Transpose -> T ,Here
# transpose makes row to columns and columns to row

t=np.array([[1,2,3],[4,5,6]])
print(t)
print(t.T)
```
